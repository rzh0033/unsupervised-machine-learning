```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
```

```python
# Load in dataset
df = pd.read_csv('crypto_data.csv', index_col = 0)
df
```

```python
df.value_counts()
```

```python
# Filter for only currently traded
filter_df = df[df['IsTrading'] == True]
filter_df
```

```python
# Isolate 'IsTrading' for later
filter_df = filter_df.drop(columns = 'IsTrading')
```

```python
filter_df
```

```python
# Data clean up time
filter_df.isnull().sum()
```

```python
filter_df = filter_df.dropna(how = 'any')
```

```python
filter_df.isnull().sum()
```

```python
# Filter for mined coins
filter_df = df[df['TotalCoinsMined'] > 0]
filter_df
```

```python
# Setup for TSNE
coinName = pd.DataFrame(filter_df['CoinName']).reset_index(drop = True)
coinName
```

```python
# Drop CoinName and Unnamed
filter_df.drop(columns = ['CoinName'], axis = 1, inplace = True)
filter_df
```

```python
X = pd.get_dummies(filter_df, columns= ['Algorithm', 'ProofType'])
X
```

```python
scaler = StandardScaler()
scale_df = scaler.fit_transform(X)
```

```python
len(X.columns)
```

```python
# PCA
pca = PCA(n_components = 4)

crypt_pca = pca.fit_transform(scale_df)
```

```python
df_pca = pd.DataFrame(data = crypt_pca, columns = ['PC 1', 'PC 2', 'PC 3', 'PC 4'])
df_pca.head()
```

```python
pca.explained_variance_ratio_
```

```python
pca_samp = PCA(n_components = .9)

crypt_pca_samp = pca_samp.fit_transform(scale_df)
crypt_pca_samp
```

```python
trans_pca_samp = pd.DataFrame(data = crypt_pca_samp)
trans_pca_samp.head()
```

```python
pca_samp.explained_variance_ratio_
```

```python
# Create plot
features = range(pca.n_components)
plt.bar(features, pca.explained_variance_ratio_, color = 'blue')
plt.xlabel('PCA Features')
plt.ylabel('Variance')
plt.xticks(features)
```

```python
#TSNE
tsne = TSNE(learning_rate = 35)
```

```python
# Reduce dimensions
tsne_features = tsne.fit_transform(trans_pca_samp)
tsne_features.shape
```

```python
X = tsne_features[:,0]
y = tsne_features[:,1]
```

```python
# Plot the results
plt.scatter(X, y)
plt.show()
```

```python
# K means
inertia = []
k = list(range(1, 11))

# Calculate the inertia
for i in k:
    km = KMeans(n_clusters = i, random_state = 0)
    km.fit(trans_pca_samp)
    inertia.append(km.inertia_)
    
elbow_data = {'k': k, 'inertia': inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.head()
```

```python
# Plot K Means
plt.plot(df_elbow['k'], df_elbow['inertia'])
plt.xticks(range(1,11))
plt.xlabel('# of Clusters')
plt.ylabel('Inertia')
plt.title('Curve for Cryptocurrency')
plt.show()
```

```python
#
# There aren't enough similarities among the coins to cluster properly 
# (there is evidence of the moving in that direction but no solid clusters to be found).
# The lack of an elbow implies that there isn't enough data to make any proper conclusion
# PCA shows 0-10 has a point of interest. When cross referenced with k-Means it seems 4-8 would be a better zone. 
# Ultimately, there is not enough solid information provided to make any solid conclusions
#
```
